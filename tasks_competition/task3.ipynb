## Question 2

Write a program that plays a familiar guessing game: first your program will set some number to be the "correct answer", then your program will ask the user to guess this answer until they get it right. For each guess your program should give a hint based on whether the guess was too high or too low.

For example, a session with your program may look like the following:

Enter your guess: 46

Your guess is too low


Enter your guess: 95

Your guess is too high

Enter your guess: 80
    
Correct you win!!!

```python
import random
random_number = random.randint(0, 100)
print(random_number)
numb1 = int(input('Enter your guess'))
while numb1 != random_number :
  if numb1< random_number:
    print('Your guess is too low')
  elif numb1 > random_number:
    print('Your guess is too high')
  numb1 = int(input('Enter your guess'))
print('Correct you win!')
```

## Question 3

Modify the program above so that the user only has 5 tries to guess. If the user is unable to guess within 5 tries, print out a statement that says you loose

```python
import random
random_number = random.randint(0, 100)
print(random_number)
numb1 = int(input('Enter your guess'))
count = 0
lost=True
while count != 5 :
  if numb1< random_number:
    print('Your guess is too low')
  elif numb1 > random_number:
    print('Your guess is too high')
  else:
      print('Correct you win!')
      lost =False
      break
  numb1 = int(input('Enter your guess'))
  count = count+1
if lost:
  print('5 chances up,you lose!')
```

## Question 5

Create a program which takes in a list of grades ( could be any size) and determines whether each grade is a fail, pass, excellent and returns the corresponding list.

Note:

* below 50% is a fail
* between 50-75% is pass
* above 75% is excellent

For example:

marks=[20,45,50,60,79]

results=[fail,fail,pass,pass,excellent]

```python
marks=[20,45,50,60,79]
results = []
for i in range(0,len(marks)):
  if marks[i]<50:
    results.append('fail')
  elif marks[i]> 50 and marks[i]<75:
    results.append('pass')
  else:
    results.append('excellent')
print(results)
```
